```python
import plotly.io as pio
import plotly.express as px
import plotly.subplots as psub
import plotly.graph_objects as go
pio.renderers.default = "notebook_connected"

import numpy as np
```

```python
import re
loss_re = re.compile(r".*loss:(\d+\.\d+)")
acc_re = re.compile(r".*accuracy: (\d+\.\d+)")
epoch_acc_re = re.compile(r".*Accuracy: (\d+\.\d+)")
```

```python
loss = []
acc = []
epoch_acc = []

with open("../cluster_results/logs_title.txt", "r") as o:
    for line in o:
        if loss_re.match(line):
            loss.append(float(loss_re.match(line).group(1)))
        if acc_re.match(line):
            acc.append(float(acc_re.match(line).group(1)))
        if epoch_acc_re.match(line):
            epoch_acc.append(float(epoch_acc_re.match(line).group(1)))
```

```python
px.line(x=range(len(loss)), y=loss, title="Loss")
```

```python
px.line(x=range(len(acc)), y=acc, title="acc")
```

```python
px.line(x=range(len(epoch_acc)), y=epoch_acc, title="epoch_acc")
```

```python
fig = psub.make_subplots(rows=3, cols=1)
fig.add_trace(go.Scatter(x=np.arange(len(loss)), y=loss, name="loss"), row=1, col=1)
fig.add_trace(go.Scatter(x=np.arange(len(acc)), y=acc, name="acc"), row=2, col=1)
fig.add_trace(go.Scatter(x=np.arange(len(epoch_acc)), y=epoch_acc, name="epoch_acc"), row=3, col=1)
fig.update_layout(height=800, width=800)
fig.show()

fig.write_html(f"a.html")
```
