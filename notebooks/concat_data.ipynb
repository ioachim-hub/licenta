```python
import pandas as pd
import numpy as np
```

### 1.	Fake News – Kaggle - https://www.kaggle.com/c/fake-news/data
    Setul de date cuprinde trei fişiere: submit.csv, test.csv, train.csv.
train.csv este folosit pentru antrenare şi test.csv pentru testare după cum
este intuitiv. Fişierul submit.csv este folosit pentru a se eticheta fiecare
ştire în parte după id după cum urmează: 1 pentru ştire falsă şi 0 pentru ştire
adevărată.
Structura fişierului de testare este următoarea : 4 coloane cu următoarele
titluri – id, titlu, autor, text. Fişierul de testare mai are în plus eticheta
pe a cincea coloană pe care o deţine structura acestuia. Pentru fişierul de
antrenare sunt 20387 intrări cu texte unice şi fişierul de testare cuprinde
5127 intrări.

```python
df_fakenews_kaggle = pd.read_csv(
    "../Datasets/httpswww.kaggle.comcfake-newsdataselect=train.csv/train.csv"
)
df_fakenews_kaggle.head(10)

df_fakenews_kaggle = df_fakenews_kaggle[["title", "text", "label"]]
df_fakenews_kaggle.head(10)
```

### 2. Fake News Detection Datasets – Universitatea din Victoria - https://www.uvic.ca/engineering/ece/isot/datasets/fake-news/index.php
    Setul de date cuprinde două fişiere: Fake.csv şi True.csv. Fiecare intrare
    din setul de date cuprinde următoarele informaţii distribuite pe căte o
    coloană: titlul articolului, textul articolului, Data în care articolul a
    fost publicat. Datele colectate sunt culese în intervalul de ani 2016-2017

```python
df_fakenews_victoria_true = pd.read_csv(
    "../Datasets/httpswww.uvic.caengineeringeceisotdatasetsfake-newsindex.php/True.csv"
)
df_fakenews_victoria_true.head(10)
df_fakenews_victoria_fake = pd.read_csv(
    "../Datasets/httpswww.uvic.caengineeringeceisotdatasetsfake-newsindex.php/Fake.csv"
)
df_fakenews_victoria_fake.head(10)
```

```python
df_fakenews_victoria_true = df_fakenews_victoria_true[["title", "text"]]
df_fakenews_victoria_true = pd.concat(
    [
        df_fakenews_victoria_true,
        pd.DataFrame(np.zeros([len(df_fakenews_victoria_true), 1]), columns=["label"]),
    ],
    axis=1,
)

df_fakenews_victoria_fake = df_fakenews_victoria_true[["title", "text"]]
df_fakenews_victoria_fake = pd.concat(
    [
        df_fakenews_victoria_fake,
        pd.DataFrame(np.ones([len(df_fakenews_victoria_fake), 1]), columns=["label"]),
    ],
    axis=1,
)

df_fakenews_victoria = pd.concat(
    [df_fakenews_victoria_true, df_fakenews_victoria_fake], axis=0
)
```

### 3.	Getting Real about Fake News – Kaggle - https://www.kaggle.com/mrisdal/fake-news
    Acest set de date este alcătuit numai din ştiri false din anul 2016 şi cuprinde
 un număr de 12999 de postări pe o perioadă de 30 de zile. Setul de date este 
 compus dintr-un fişier numit fake.csv care are 12999 intrări organizate pe 10 
 coloane în această ordine: uuid, gradul de ameninţare, autor, date de publicare, 
 titlul, text, limbă, data la care a fost colectată ştirea, url-ul site-ului, 
 ţara.

```python
df_fakenews_just = pd.read_csv("../Datasets/httpswww.kaggle.commrisdalfake-news/fake.csv")
df_fakenews_just.head(10)

df_fakenews_just = df_fakenews_just[["title", "text"]]
df_fakenews_just = pd.concat(
    [
        df_fakenews_just,
        pd.DataFrame(np.ones([len(df_fakenews_just), 1]), columns=["label"]),
    ],
    axis=1,
)
df_fakenews_just.head(10)
```

### 4.	Fake News Dataset – Kaggle - https://www.kaggle.com/mohamadalhasan/a-fake-news-dataset-around-the-syrian-war
    Articolele colectate au fost etichetate de către o grupare de 
experţi din anumite site uri ţintă cunoscute drept furnizoare de 
ştiri false. Etichete care au fost atribuite ştirilor sunt 
următoarele: 0 pentru ştiri false şi 1 pentru ştiri credibile.
    Setul de date este alcătuit din 804 articole echitetate ca 
adevărate sau false care sunt structurate pe 7 coloane după cum 
urmează: id, titlul articolului, conţinutul articolului, sursa din 
care a fost colectat articolul, data în care a fost scris, locatia 
şi eticheta de adevărat sau fals pe care o deţine.

```python
df_fakenews_dataset_kaggle = pd.read_csv(
    "../Datasets/httpswww.kaggle.commohamadalhasana-fake-news-dataset-around-the-syrian-war/FA-KES-Dataset.csv", encoding= 'unicode_escape'
)
df_fakenews_dataset_kaggle.head(10)

df_fakenews_dataset_kaggle = df_fakenews_dataset_kaggle[
    ["article_title", "article_content", "labels"]
]
df_fakenews_dataset_kaggle.columns = ["title", "text", "label"]

df_fakenews_dataset_kaggle = df_fakenews_dataset_kaggle.replace({0: 1, 1: 0})
df_fakenews_dataset_kaggle.head(10)
```

### 5.	Fake News – Kaggle - https://www.kaggle.com/hassanamin/textdb3
   Setul de date cuprinde 7796 intrări structurate pe 4 coloane 
după cum urmează: id, titlu, conţinutul articolului şi eticheta pe 
care a primit-o de TRUE pentru ştiri adevărate sau FAKE pentru ştiri 
false.

```python
df_fakenews_db3 = pd.read_csv(
    "../Datasets/httpswww.kaggle.comhassanamintextdb3/fake_or_real_news.csv"
)
df_fakenews_db3.head(10)

df_fakenews_db3 = df_fakenews_db3[["title", "text", "label"]]

df_fakenews_db3 = df_fakenews_db3.replace("FAKE", 1)
df_fakenews_db3 = df_fakenews_db3.replace("REAL", 0)
```

### 6. FakeNewsDataset

```python
missing_files_list = []

df_text = pd.read_csv("../FakeNewsNet/code/dataset_text.csv")
df_title = pd.read_csv("../FakeNewsNet/code/dataset_title.csv")

text = df_text[['text']]
title = df_title[['text']]
label = df_text[['label']]
df_fakenews_net = pd.concat([title, text, label], axis=1)
df_fakenews_net.columns = ["title", "text", "label"]
```

## Concatenare

```python
df_total = pd.concat(
    [
        df_fakenews_kaggle,
        df_fakenews_victoria,
        df_fakenews_just,
        df_fakenews_dataset_kaggle,
        df_fakenews_db3,
        df_fakenews_net,
    ],
    axis=0,
)

df_total = df_total.fillna("News")

df_total["index"] = range(1, len(df_total) + 1)
df_total = df_total.set_index("index")

df_total.to_csv("Raw_Dataset.csv")
```
