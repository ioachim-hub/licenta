```python
from requests_html import HTMLSession
```

```python
session = HTMLSession()
```

```python
url = "https://www.biziday.ro/feed/"
```

```python
data = session.get(url)
```

```python
data.html.raw_html
```

```python
import xml.etree.ElementTree as ET
from xml.dom.minidom import parse, parseString
import xmltodict
import pandas as pd
import bs4
```

```python
root = xmltodict.parse(str(data.html.raw_html, encoding="utf-8"))
```

```python

for item in root['rss']['channel']['item']:
    text = ""
    soup = bs4.BeautifulSoup(item["content:encoded"])
```

```python
for item in root['rss']['channel']['item']:
    print(f"Title is: {item['title']}")
    print(f"Link is: {item['link']}")
    print(f"Pub date is: {pd.to_datetime(item['pubDate'])}")
    print(f"Description is: {bs4.BeautifulSoup(item["content:encoded"]).get_text()}")
```
